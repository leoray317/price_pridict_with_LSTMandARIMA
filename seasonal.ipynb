```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('month_data_stata.csv')
```

```python
d=0

for i in range(1999,2019):
    year=[]
    if df.year.values[d] == i:
        year.append(df.price.values[d])
    d+=1
    plt.plot(year)
    plt.show()
```

```python
print([str(x) for x in range(1999,2020)])
```

```python
year_data=[]
for i in range(1999,2020):
    d=0
    year=[]
    for j in df.year.values:
        if j == i:
            year.append(df.price.values[d])
        d+=1
    year_data.append(year)

pd.DataFrame(year_data,columns=[str(x) for x in range(1,13)])
```

```python
for i in range(21):
    plt.plot(year_data[i])
plt.savefig('seasonal.png')
```
